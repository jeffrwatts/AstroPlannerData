```python
# Cell 5 — Data source lookup (OpenNGC)
# Fetches RA, Dec, constellation, subType, magnitude, and angular size from OpenNGC CSV.
# Raw lookup results only — overrides are not applied here.

def parse_ra_degrees(ra_str):
    """Convert RA from HH:MM:SS.ss to decimal degrees."""
    if not ra_str or ra_str.strip() == "":
        return None
    parts = ra_str.strip().split(":")
    h, m, s = float(parts[0]), float(parts[1]), float(parts[2])
    return round((h + m / 60 + s / 3600) * 15, 7)


def parse_dec_degrees(dec_str):
    """Convert Dec from +DD:MM:SS.ss to decimal degrees."""
    if not dec_str or dec_str.strip() == "":
        return None
    sign = -1 if dec_str.strip().startswith("-") else 1
    parts = dec_str.strip().lstrip("+-").split(":")
    d, m, s = float(parts[0]), float(parts[1]), float(parts[2])
    return round(sign * (d + m / 60 + s / 3600), 7)


def object_id_to_opengc_name(object_id):
    """Convert an objectId to the OpenNGC Name field format.
    Examples: ngc869 → NGC0869, ic434 → IC0434
    Returns None for Messier, Sharpless, Barnard, and custom IDs.
    """
    oid = object_id.lower()
    if oid.startswith("ngc"):
        return f"NGC{int(oid[3:]):04d}"
    elif oid.startswith("ic"):
        return f"IC{int(oid[2:]):04d}"
    return None


# OpenNGC type codes → clean subType strings
OPENGC_TYPE_MAP = {
    # Galaxies
    "G":      "Spiral Galaxy",
    "GX":     "Spiral Galaxy",
    "E":      "Elliptical Galaxy",
    "S":      "Spiral Galaxy",
    "S0":     "Lenticular Galaxy",
    "SA":     "Spiral Galaxy",
    "SAB":    "Spiral Galaxy",
    "SB":     "Barred Spiral Galaxy",
    "Irr":    "Irregular Galaxy",
    "dE":     "Dwarf Galaxy",
    "GGroup": "Spiral Galaxy",
    "GPair":  "Spiral Galaxy",
    "GTrpl":  "Spiral Galaxy",
    # Nebulae
    "EN":     "Emission Nebula",
    "RN":     "Reflection Nebula",
    "PN":     "Planetary Nebula",
    "SNR":    "Supernova Remnant",
    "DN":     "Dark Nebula",
    "HII":    "HII Ionized region",
    "ER":     "Emission Nebula",     # emission/reflection → emission
    "EmN":    "Emission Nebula",
    "RfN":    "Reflection Nebula",
    "Neb":    "Emission Nebula",     # generic nebula — override individually if reflection
    # Clusters
    "OC":     "Open Cluster",
    "OCl":    "Open Cluster",
    "GC":     "Globular Cluster",
    "GCl":    "Globular Cluster",
    "Cl+N":   "Open Cluster",
    "SC":     "Star Cloud",
    "Ast":    "Asterism",
    "*Ass":   "Star Cloud",          # stellar association (e.g. M24 Sagittarius Star Cloud)
    "Other":  "Asterism",            # catch-all for miscellaneous (e.g. M73)
    # Stars
    "DS":     "Double Star",
    "*":      "Double Star",
    "**":     "Double Star",
    # Duplicate entries — no type; handled via subType override in Cell 4
    "Dup":    None,
}


# Load OpenNGC CSV
with urllib.request.urlopen(OPENGC_CSV_URL) as response:
    csv_text = response.read().decode("utf-8")

opengc_df = pd.read_csv(io.StringIO(csv_text), sep=";", dtype=str, keep_default_na=False)
opengc_df.columns = opengc_df.columns.str.strip()

# Primary lookup: NGC/IC name → row (e.g. "NGC0869")
opengc_lookup = {row["Name"].strip(): row for _, row in opengc_df.iterrows()}

# Secondary lookup: Messier number (int) → row via the "M" column
messier_lookup = {}
for _, row in opengc_df.iterrows():
    m_val = row.get("M", "").strip()
    if m_val:
        try:
            messier_lookup[int(m_val)] = row
        except ValueError:
            pass

print(f"OpenNGC loaded: {len(opengc_df)} entries")
print(f"Messier entries indexed: {len(messier_lookup)}")
```

```python
# Cell 5c — SIMBAD fallback lookup
# Used for any objectId that OpenNGC cannot resolve (Sharpless, Barnard, LDN, etc.)

import time

# Only request otype — adding magnitude fields (V, B) causes SIMBAD to drop
# objects that have no magnitude entry (most nebulae), returning zero results.
_simbad = Simbad()
_simbad.add_votable_fields("otype")
_simbad.TIMEOUT = 20   # seconds per request

# Map SIMBAD object type codes → our subType strings
SIMBAD_TYPE_MAP = {
    # Nebulae
    "HII":    "Emission Nebula",
    "EmO":    "Emission Nebula",
    "Neb":    "Emission Nebula",
    "WR*":    "Emission Nebula",    # Wolf-Rayet bubble (e.g. Sh2-308)
    "RNe":    "Reflection Nebula",
    "DNe":    "Dark Nebula",
    "MolCld": "Dark Nebula",
    "CGb":    "Dark Nebula",
    "PN":     "Planetary Nebula",
    "SNR":    "Supernova Remnant",
    "ISM":    "Emission Nebula",
    # Galaxies
    "G":      "Spiral Galaxy",
    "AGN":    "Spiral Galaxy",
    "GiG":    "Irregular Galaxy",
    "GiP":    "Spiral Galaxy",
    "SyG":    "Spiral Galaxy",
    "GGroup": "Spiral Galaxy",
    "GPair":  "Spiral Galaxy",
    # Clusters
    "GlC":    "Globular Cluster",
    "OpC":    "Open Cluster",
    "Cl*":    "Open Cluster",
    "Assoc*": "Open Cluster",
    # Stars
    "**":     "Double Star",
    "*":      "Double Star",
}


def objectid_to_simbad_name(object_id):
    """Convert an objectId to a SIMBAD-resolvable identifier string."""
    oid = object_id.lower()
    m = re.match(r'^sh2-(\d+)$', oid)
    if m:
        return f"Sh 2-{m.group(1)}"
    m = re.match(r'^barnard(\d+)$', oid)
    if m:
        return f"Barnard {m.group(1)}"
    m = re.match(r'^ldn(\d+)$', oid)
    if m:
        return f"LDN {m.group(1)}"
    return object_id


def simbad_lookup(object_id):
    """Query SIMBAD for a single object. Returns a result dict or None."""
    name = objectid_to_simbad_name(object_id)
    try:
        result = _simbad.query_object(name)
    except Exception as e:
        print(f"  SIMBAD error for {name!r}: {e}")
        return None
    if result is None or len(result) == 0:
        return None

    row = result[0]

    try:
        ra = float(row["ra"])
    except Exception:
        ra = None
    try:
        dec = float(row["dec"])
    except Exception:
        dec = None

    raw_type = str(row["otype"]).strip() if row["otype"] else ""
    sub_type = SIMBAD_TYPE_MAP.get(raw_type)

    return {
        "objectId":         object_id,
        "ra":               ra,
        "dec":              dec,
        "constellation":    None,
        "subType":          sub_type,
        "magnitude":        None,
        "angularSizeMajor": None,
        "angularSizeMinor": None,
        "_simbad_name":     name,
        "_simbad_type":     raw_type,
    }


print("SIMBAD fallback configured.")
```

```python
# Cell 5b — Perform lookups

# Aliases: objectId → NGC/IC name for objects whose Messier number isn't in OpenNGC's M column.
OPENGC_NAME_ALIASES = {
    "m102": "NGC5866",   # M102 conventionally = NGC 5866 (Spindle Galaxy)
}

# Manual data: last-resort fallback for objects that can't be resolved programmatically.
# RA/Dec in decimal degrees. Angular sizes in arcminutes.
MANUAL_DATA = {
    # Leo Triplete — non-standard objectId
    "leotriplet": {
        "ra": 169.963, "dec": 13.280, "constellation": "Leo",
        "subType": "Galaxy Group", "magnitude": None,
        "angularSizeMajor": 48.0, "angularSizeMinor": 30.0,
    },
    # Rho Oph — non-standard objectId, SIMBAD won't match "rhocomplex"
    "rhocomplex": {
        "ra": 246.396, "dec": -23.45, "constellation": "Oph",
        "subType": "Emission Nebula", "magnitude": None,
        "angularSizeMajor": 270.0, "angularSizeMinor": 270.0,
    },
}


def extract_row_data(row, oid):
    ra  = parse_ra_degrees(row.get("RA", ""))
    dec = parse_dec_degrees(row.get("Dec", ""))

    raw_type = row.get("Type", "").strip()
    sub_type = OPENGC_TYPE_MAP.get(raw_type)

    hubble = row.get("Hubble", "").strip()
    if sub_type in ("Spiral Galaxy", "Elliptical Galaxy", "Lenticular Galaxy") and hubble:
        if hubble.startswith("SB"):
            sub_type = "Barred Spiral Galaxy"
        elif hubble.startswith("S0") or hubble == "S0":
            sub_type = "Lenticular Galaxy"
        elif hubble.startswith("S"):
            sub_type = "Spiral Galaxy"
        elif hubble.startswith("E"):
            sub_type = "Elliptical Galaxy"
        elif hubble.startswith("Irr") or hubble.startswith("I"):
            sub_type = "Irregular Galaxy"

    try:
        mag = float(row.get("V-Mag") or row.get("B-Mag") or "")
    except (ValueError, KeyError):
        mag = None

    try:
        size_major = float(row.get("MajAx", ""))
    except (ValueError, KeyError):
        size_major = None

    try:
        size_minor = float(row.get("MinAx", ""))
    except (ValueError, KeyError):
        size_minor = None

    constellation = row.get("Const", "").strip() or None

    return {
        "objectId":         oid,
        "ra":               ra,
        "dec":              dec,
        "constellation":    constellation,
        "subType":          sub_type,
        "magnitude":        mag,
        "angularSizeMajor": size_major,
        "angularSizeMinor": size_minor,
    }


lookup_results = []
not_found = []
simbad_hits = []

for obj in objects:
    oid = obj["objectId"]
    row = None

    # 1. OpenNGC by NGC/IC name
    opengc_name = object_id_to_opengc_name(oid)
    if opengc_name and opengc_name in opengc_lookup:
        row = opengc_lookup[opengc_name]

    # 2. OpenNGC by Messier number
    if row is None:
        oid_lower = oid.lower()
        if oid_lower.startswith("m") and oid_lower[1:].isdigit():
            row = messier_lookup.get(int(oid_lower[1:]))

    # 3. OpenNGC via alias (e.g. m102 → NGC5866)
    if row is None:
        alias = OPENGC_NAME_ALIASES.get(oid.lower())
        if alias:
            alias_name = object_id_to_opengc_name(alias) or alias
            row = opengc_lookup.get(alias_name)

    if row is not None:
        lookup_results.append(extract_row_data(row, oid))
        continue

    # 4. SIMBAD fallback — small delay between requests to avoid rate limiting
    time.sleep(0.5)
    print(f"  SIMBAD: querying {oid}...")
    result = simbad_lookup(oid)
    if result is not None:
        simbad_hits.append(f"{oid} → {result['_simbad_name']} ({result['_simbad_type']})")
        result.pop("_simbad_name", None)
        result.pop("_simbad_type", None)
        lookup_results.append(result)
        continue

    # 5. Manual data
    if oid.lower() in MANUAL_DATA:
        lookup_results.append({"objectId": oid, **MANUAL_DATA[oid.lower()]})
        continue

    # 6. Not found
    not_found.append(oid)
    lookup_results.append({"objectId": oid, "ra": None, "dec": None,
                            "constellation": None, "subType": None,
                            "magnitude": None, "angularSizeMajor": None,
                            "angularSizeMinor": None})

print(f"\nLooked up: {len(lookup_results) - len(not_found)}/{len(objects)}")
if simbad_hits:
    print(f"\nSIMBAD resolved {len(simbad_hits)} objects:")
    for h in simbad_hits:
        print(f"  {h}")
if not_found:
    print(f"\nNot found anywhere: {not_found}")
```

```python
# Cell 6 — Build & validate DataFrame

# Build base DataFrame from object list
base_df = pd.DataFrame(objects)

# Merge with lookup results
lookup_df = pd.DataFrame(lookup_results)
df = base_df.merge(lookup_df, on="objectId", how="left")

# Apply subType overrides
for oid, override_subtype in SUBTYPE_OVERRIDES.items():
    mask = df["objectId"] == oid
    df.loc[mask, "subType"] = override_subtype

# Derive type from subType
df["type"] = df["subType"].map(SUBTYPE_TO_TYPE)

# Expand constellation abbreviations to full names
df["constellation"] = df["constellation"].map(
    lambda abbr: CONSTELLATION_NAMES.get(abbr, abbr) if pd.notna(abbr) else None
)

# ── Validation ──
errors = []

# 1. Missing RA/Dec
missing_coords = df[df["ra"].isna() | df["dec"].isna()]["objectId"].tolist()
if missing_coords:
    errors.append(f"Missing ra/dec for: {missing_coords}")

# 2. Unmapped subTypes
unmapped = df[df["type"].isna() & df["subType"].notna()][["objectId", "subType"]].drop_duplicates()
if not unmapped.empty:
    errors.append(f"Unmapped subTypes — add to SUBTYPE_TO_TYPE:\n{unmapped.to_string(index=False)}")

# 3. Null subType (not found anywhere and no override)
null_subtype = df[df["subType"].isna()]["objectId"].tolist()
if null_subtype:
    errors.append(f"No subType resolved (not in OpenNGC, no override): {null_subtype}")

# 4. Duplicate objectIds
dupes = df[df.duplicated("objectId")]["objectId"].tolist()
if dupes:
    errors.append(f"Duplicate objectIds: {dupes}")

if errors:
    raise ValueError("Validation failed:\n" + "\n".join(errors))

print(f"Validation passed. {len(df)} objects ready.")
df[["displayName", "objectId", "type", "subType", "constellation", "ra", "dec",
    "angularSizeMajor", "angularSizeMinor", "magnitude", "recommended"]]
```
